```python
import pandas as pd
```

```python
# Create input data

TLD = ["io", "com",'tech', 'ai','eu', 'earth', 'dev']
SLD = ['AutomateSpatial', 'QuantGIS', 'GeoPipeline','SpatialAI', 'MapStack', 'SpatialFlow', 'MapLogic', 
       'VertexAI' , "GeoQuant", 'SpaceAI', 'SpaceFlow', 'QuantSpatial' , 'SpatialFlow', 'SpatialAI', 'GeoQuant', 
      'SpatialQuant']
```

```python
# Save input data into files rte
def saveInput(fileName, InputData):
    with open(f"Data/{fileName}.csv", mode="w") as f:
        f.write(",".join(InputData)) # (converts into comma separated string)

saveInput("TLD", TLD)
saveInput("SLD", SLD)
```

```python
# Concatenate  SLD with TLD

concatenated = []

for sld_element in SLD:
    concatenated.extend([sld_element + "." + tld_element for tld_element in TLD ])
print(concatenated)
```

```python
# Convert to Pandas Series and save to CSV

webPages = pd.Series(concatenated)
webPages.to_csv("Data/webPages.csv", header=False, index = False)
```
